```python
cols = [' alcalinity_of_ash',' magnesium',' flavanoids',
        ' color_intensity',' proline' ]
```

```python
X = X[cols]
```

```python
X.head()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
isf = IsolationForest(n_estimators=100, 
                      random_state=0, contamination=0.1)
```

```python
isf = isf.fit(X)
clean = isf.predict(X)  # Outlier detection
```

```python
clean
```

```python
X = X[clean==1]  # Boolean filtering, removal
y = y[clean==1]
```

```python
X.shape
```

```python
import seaborn as sns
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
rus = RandomOverSampler(random_state = 0) 
```

```python
X_res, y_res = rus.fit_resample(X, y)
```

```python
X_res = pd.DataFrame(X_res) 
Y_res = pd.DataFrame(y_res) 
```

```python
Y_res.value_counts()
```

```python
X_res.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_s = scaler.fit_transform(X_res)
pd.DataFrame(X_s).describe()
```

```python
# Cross validation
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_s, y_res, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#classifier = KNeighborsClassifier(n_neighbors=5)
classifier = MLPClassifier(hidden_layer_sizes=(50,50,50),
                          max_iter=100,
                          random_state=0)
```

```python
classifier.fit(X_train, y_train) # training
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) * 100
```

```python
import joblib 
joblib.dump(classifier, 'wine.model') 
```

```python
joblib.dump(scaler, 'scaler.model', compress=9) 
```
